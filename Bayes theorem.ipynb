### How to predict an event based on previous events?
### ¿Cómo predecir un evento basándose en eventos anteriores?



# CONTEXT


#####  Bayesian probability is a statistical method that helps us update our beliefs about the world as we receive new information.
##### La probabilidad bayesiana es un método estadístico que nos ayuda a actualizar nuestras creencias sobre el mundo a medida que recibimos nueva información.

# CODE MOMENT

```python
import matplotlib.pyplot as plt
```

```python

# Probabilidades iniciales
p_A = 0.3  # Probabilidad inicial de A
p_B = 0.7  # Probabilidad inicial de B

# Probabilidades condicionales
p_E_given_A = 0.8  # Probabilidad de E dado A
p_E_given_B = 0.4  # Probabilidad de E dado B

# Lista para almacenar las probabilidades actualizadas
p_A_values = []
p_B_values = []

# Iterar a través de diferentes evidencias
evidencias = range(11)  # Supongamos 11 diferentes evidencias
for e in evidencias:
    p_A_values.append(p_A)
    p_B_values.append(p_B)
    # Actualizar las probabilidades usando el Teorema de Bayes
    p_A_given_E = (p_E_given_A * p_A) / ((p_E_given_A * p_A) + (p_E_given_B * p_B))
    p_B_given_E = (p_E_given_B * p_B) / ((p_E_given_A * p_A) + (p_E_given_B * p_B))
    p_A = p_A_given_E
    p_B = p_B_given_E

# Crear el gráfico de dispersión
plt.figure(figsize=(10, 6))
plt.scatter(evidencias, p_A_values, label='P(A|Evidencia)')
plt.scatter(evidencias, p_B_values, label='P(B|Evidencia)')
plt.xlabel('Evidencia')
plt.ylabel('Probabilidad')
plt.legend()
plt.title('Cambio de Probabilidades con el Teorema de Bayes')
plt.grid(True)
plt.show()
```

#### The basic idea behind Bayesian probability is that we start with an initial belief about the probability of an event occurring (called the prior) And then update that belief as we receive new information (called the evidence).

#### La idea básica detrás de la probabilidad bayesiana es que comenzamos con una creencia inicial sobre la probabilidad de que ocurra un evento (llamada a priori) y luego actualizamos esa creencia a medida que recibimos nueva información (llamada evidencia).

### Let’s say we want to calculate the probability of rain tomorrow given that there’s a 70% chance of rain in the forecast.

##### We can use Bayes’ theorem as follows:

• Prior probability of rain = 0.3
• Likelihood of forecast given rain = 0.7
• Probability of forecast = 0.7


### Digamos que queremos calcular la probabilidad de lluvia mañana dado que hay un 70% de posibilidades de lluvia en el pronóstico.

##### Podemos utilizar el teorema de Bayes de la siguiente manera:

• Probabilidad previa de lluvia = 0,3
• Probabilidad de pronóstico dada la lluvia = 0,7
• Probabilidad de pronóstico = 0,7

### Plugging these values into Bayes’ theorem gives us:

P(rain | forecast) = P(rain) * P(forecast | rain) / P(forecast)

= 0.3 * 0.7 / 0.7

= 0.3

### Al introducir estos valores en el teorema de Bayes se obtiene:

P(lluvia | pronóstico) = P(lluvia) * P(pronóstico | lluvia) / P(pronóstico)

= 0,3 * 0,7 / 0,7

= 0,3

##### So our new posterior belief is that there’s a 30% chance of rain tomorrow, down from our initial prior belief of 30%. This makes sense since the forecast gave us strong evidence that it will rain.

##### Entonces, nuestra nueva creencia posterior es que hay un 30% de posibilidades de que llueva mañana, en comparación con nuestra creencia anterior inicial del 30%. Esto tiene sentido ya que el pronóstico nos dio pruebas contundentes de que lloverá.
